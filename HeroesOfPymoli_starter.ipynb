```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remem#ber to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data= pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Total number of players 
player= len(purchase_data["SN"].value_counts())
#storing number counts in a column
player_count=pd.DataFrame([player], columns = ["Total Players"])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Summary Data Calculations
#finding unique number of purchase items
unique_purchase_items = len(purchase_data["Item Name"].unique())
#finding the average
average_price= purchase_data["Price"].mean()
#finding the revenue
revenue= purchase_data["Price"].sum()
#finding purchase items
number_purchase_items= len(purchase_data["Item Name"])
#summary
summary_data=pd.DataFrame({"Number of Unique Items": [unique_purchase_items],
                           "Average Price":[average_price],
                           "Number of Purchase Items": [number_purchase_items],
                           "Revenue":[revenue]})
summary_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#grouping by gender
gender =purchase_data.groupby("Gender")
#unique gender count
gender_count= gender.nunique()["SN"]
#percentage of players
percentage_players=gender_count/player *100
#creating a dataframe
gender_demographics_df=pd.DataFrame({"Total Count":gender_count,"Percentage of Players":percentage_players})
gender_demographics_df.index.name=None

#formatting
gender_demographics_df.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Count the total purchases by gender 
purchase_count = gender["Purchase ID"].count()

#average purchase price
average_purchase_price=gender["Price"].mean()

total_price=gender["Price"].sum()

# Average purchase total by gender divivded by purchase count by unique shoppers
average_purchase_per_person = total_price/gender_count

# Gender Summary Dataframe
gender_demographics_df = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": average_purchase_price,
                                    "Average Purchase Value":total_price,
                                    "Average Purchase Total per Person": average_purchase_per_person})
# Provide index in top left as "Gender"
gender_demographics_df.index.name = "Gender"

# Format with currency style 
gender_demographics_df["Average Purchase Value"]=gender_demographics_df["Average Purchase Value"].map("${:.2f}".format)
gender_demographics_df["Average Purchase Price"]=gender_demographics_df["Average Purchase Price"].map("${:.2f}".format)
gender_demographics_df["Average Purchase Total per Person"]=gender_demographics_df["Average Purchase Total per Person"].map("${:.2f}".format)

gender_demographics_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Creating Bins
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.0, 39.9, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorizing the existing players using the age bins
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names, include_lowest=True)

#Group by Age
age_group= purchase_data.groupby("Age Group")

#Counting unique numbers in the Age group
total_count_age_group= age_group["SN"].nunique()

#Calculating Perecentage of Players
    #total unique count by age/total players calculated in the first section
percentage_players= (total_count_age_group/player)*100

#Creating Age Demographic DataFrame 
age_demographics_df= pd.DataFrame({"Total Count": total_count_age_group,"Percentage of Players":percentage_players})

#Format the data frame with no index name in the corner
age_demographics_df.index.name = None

#Formating Age Demographics
age_demographics_df["Percentage of Players"]=age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

age_demographics_df
```

```python
#Using the group by age bins DataFrame

#Counting purchases by age group
purchase_count_age= age_group["Purchase ID"].count()
#Average purchase price
average_purchase_price_age= age_group["Price"].mean()
#Total purchase price
total_purchase_price=age_group["Price"].sum()

#Calculating average purchase total per person
average_total_per_person= total_purchase_price/total_count_age_group

#Creating DataFrames
purchase_analysis_age_df=pd.DataFrame({"Purchase Count": purchase_count_age,"Average Purchase Price": average_purchase_price_age,
                                  "Total Purchase Value": total_purchase_price,"Avg Total per Price":average_total_per_person})
#Format
purchase_analysis_age_df["Total Purchase Value"]=purchase_analysis_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_age_df["Average Purchase Price"]=purchase_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_age_df["Avg Total per Price"]=purchase_analysis_age_df["Avg Total per Price"].map("${:.2f}".format)

purchase_analysis_age_df.index.name=None
purchase_analysis_age_df
```

## Top Spenders

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group purhcases by players
spender_group=purchase_data.groupby("SN")

#Counting the total purhcases by player
purchase_count_spender=spender_group["Purchase ID"].count()

#Calculating the average purchase by player
average_purchase_price_spender=spender_group["Price"].mean()

#Purchase Total
purchase_total= spender_group["Price"].sum()

#Create Data frame
top_spenders_df=pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price":average_purchase_price_spender,
                             "Total Purchase Value": purchase_total})

#Sort descending order
top_spenders_df.sort_values(["Total Purchase Value"],ascending=False).head()

#top_spenders_df["Average Purchase Price"]=top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
#top_spenders_df["Total Purchase Value"]=top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieving the Item ID, Item Name, and Item Price 
items = purchase_data[["Item ID","Item Name","Price"]]

#Grouping items
items_data=items.groupby(["Item ID","Item Name"])

#Counting the number of times an item has been purchased
purchase_item_count= items_data["Price"].count()

#Calculating the purchase value per item
total_purchase=(items_data["Price"].sum())

#Item price
item_price=total_purchase/purchase_item_count

#Dataframe
most_popular_items = pd.DataFrame({"Purchase Count": purchase_item_count, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":total_purchase})

# Sort in descending order to obtain top spender names and provide top 5 item names
popular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()

# Format with currency style
popular_formatted.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
